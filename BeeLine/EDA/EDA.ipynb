```python
# 결측치를 처리할 col들의 데이터 분포 확인

# 평균기온 분포
plt.figure(figsize=(10, 6))
sns.histplot(df['평균기온(°C)'], kde=True, bins=30)
plt.show()

# 일강수량 분포
plt.figure(figsize=(10, 6))
sns.histplot(df['일강수량(mm)'], kde=True, bins=30)
plt.show()

# 평균 풍속 분포
plt.figure(figsize=(10, 6))
sns.histplot(df['평균 풍속(m/s)'], kde=True, bins=30)
plt.show()

# 골프장, 경지정리가 된 논, 경지정리가 안 된 논의 분포
plt.figure(figsize=(10, 6))
sns.histplot(df['골프장'], kde=True, bins=30)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['경지정리가 된 논'], kde=True, bins=30)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['경지정리가 안 된 논'], kde=True, bins=30)
plt.show()
```

```python
# 기온, 강수량, 풍속의 결측값 처리 -> 평균으로

# 결측값 대체 전후의 분포 시각화
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
columns = ['평균기온(°C)', '일강수량(mm)', '평균 풍속(m/s)']

for i, column in enumerate(columns):
    sns.histplot(df[column].dropna(), kde=True, ax=axes[0, i])
    axes[0, i].set_title(f'{column} (Before Imputation)')

    # 결측값 대체 후
    temp_data = df[column].copy()
    temp_data.fillna(temp_data.mean(), inplace=True)
    sns.histplot(temp_data, kde=True, ax=axes[1, i])
    axes[1, i].set_title(f'{column} (After Imputation)')

plt.tight_layout()
plt.show()
```

```python
# 골프장, 경지정리가 된 논, 경지정리가 안 된 논의 결측치 -> 0으로 (평균보다 원본데이터와 유사)

# 결측값을 0으로 대체한 데이터
data_zero_filled = df.fillna(0)

# 결측값을 각 열의 평균으로 대체한 데이터
data_mean_filled = df.copy()
data_mean_filled['골프장'].fillna(data_mean_filled['골프장'].mean(), inplace=True)
data_mean_filled['경지정리가 된 논'].fillna(data_mean_filled['경지정리가 된 논'].mean(), inplace=True)
data_mean_filled['경지정리가 안 된 논'].fillna(data_mean_filled['경지정리가 안 된 논'].mean(), inplace=True)

# 분포 시각화
fig, axes = plt.subplots(3, 3, figsize=(12, 18))

sns.histplot(data=df, x='골프장', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('골프장')
sns.histplot(data=data_zero_filled, x='골프장', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('골프장 (결측값 0으로 대체)')
sns.histplot(data=data_mean_filled, x='골프장', kde=True, ax=axes[0, 2])
axes[0, 2].set_title('골프장 (결측값 평균으로 대체)')


sns.histplot(data=df, x='경지정리가 된 논', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('경지정리가 된 논')
sns.histplot(data=data_zero_filled, x='경지정리가 된 논', kde=True, ax=axes[1, 1])
axes[1, 1].set_title('경지정리가 된 논 (결측값 0으로 대체)')
sns.histplot(data=data_mean_filled, x='경지정리가 된 논', kde=True, ax=axes[1, 2])
axes[1, 2].set_title('경지정리가 된 논 (결측값 평균으로 대체)')


sns.histplot(data=df, x='경지정리가 안 된 논', kde=True, ax=axes[2, 0])
axes[2, 0].set_title('경지정리가 안 된 논')
sns.효(data=data_zero_filled, x='경지정리가 안 된 논', kde=True, ax=axes[2, 1])
axes[2, 1].set_title('경지정리가 안 된 논 (결측값 0으로 대체)')
sns.histplot(data=data_mean_filled, x='경지정리가 안 된 논', kde=True, ax=axes[2, 2])
axes[2, 2].set_title('경지정리가 안 된 논 (결측값 평균으로 대체)')

plt.tight_layout()
plt.show()
```

```python
# Pairplot을 통한 변수 간 관계 시각화
sns.pairplot(df[['평균기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '골프장', '경지정리가 된 논', '경지정리가 안 된 논']])
plt.show()

# 상관 행렬 히트맵
plt.figure(figsize=(12, 8))
correlation_matrix = df[['평균기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '골프장', '경지정리가 된 논', '경지정리가 안 된 논']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import geopandas as gpd
vegetarian = gpd.read_file("/content/drive/MyDrive/Dart-B/환파주의보/data/2020seoulcityecology_data/현장조사.shp")
```

```python
vegetarian
```

```python
vegetarian = vegetarian[["자치구", "현존식생", "면적"]]
vegetarian = pd.DataFrame(vegetarian)
vegetarian = vegetarian[vegetarian["현존식생"].str.contains('B') | vegetarian["현존식생"].str.contains('D')]
vegetarian = vegetarian[vegetarian["현존식생"].str.startswith(('B', 'D'))]
# 열 이름 바꾸기
vegetarian = vegetarian.rename(columns = {"자치구" : "구"})
vegetarian
```

```python
def remove_prefix_and_numbers(value):
    if value[0].isalpha():
        return value[0]
    return value

# Apply the function to the DataFrame
vegetarian["현존식생"] = vegetarian["현존식생"].apply(remove_prefix_and_numbers)
# groupby(구)
vegetarian = vegetarian.groupby(["구", "현존식생"]).sum().reset_index()
vegetarianB = vegetarian[vegetarian["현존식생"] == "B"]
vegetarianB = vegetarianB.drop("현존식생", axis = 1)
vegetarianD = vegetarian[vegetarian["현존식생"] == "D"]
vegetarianD = vegetarianD.drop("현존식생", axis = 1)
```

```python
vegetarian_total = vegetarian.groupby(["구"]).sum().reset_index()
```

```python
vegetarian_total
```

```python
# 현존식생 타입별(B, D)과 총

fig, axes = plt.subplots(3, 1, figsize=(12, 12))

sns.barplot(x='구', y='면적', data=vegetarianB, ax=axes[0])
axes[0].set_title('구별 B 유형 현존식생 면적')
axes[0].tick_params(axis='x', rotation=90)

sns.barplot(x='구', y='면적', data=vegetarianD, ax=axes[1])
axes[1].set_title('구별 D 유형 현존식생 면적')
axes[1].tick_params(axis='x', rotation=90)

sns.barplot(x='구', y='면적', data=vegetarian_total, ax=axes[2])
axes[2].set_title('구별 총합 현존식생 면적')
axes[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
# 현존식생 면적이 높은 3곳과 낮은 3곳의 차이
```

```python
# 현존식생 면적이 높은 3곳
highest_three = vegetarian_total.nlargest(3, '면적')['구']

# 현존식생 면적이 낮은 3곳
lowest_three = vegetarian_total.nsmallest(3, '면적')['구']
```

```python
col_list = ['평균기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '골프장', '경지정리가 된 논', '경지정리가 안 된 논']

import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(col_list), ncols=2, figsize=(15, 15))

# Iterate through the columns of df_highest_three
for i, col in enumerate(col_list):
    # Plot the data for the current column
    sns.histplot(data=df_highest_three, x=col, kde=True, ax=axes[i,0])
    axes[i, 0].set_title(f'highest_{col}')
    sns.histplot(data=df_lowest_three, x=col, kde=True, ax=axes[i,1])
    axes[i, 1].set_title(f'lowest_{col}')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
